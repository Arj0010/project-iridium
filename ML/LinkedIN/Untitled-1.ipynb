```python
missing_summary = df.isnull().sum()
missing_summary = missing_summary[missing_summary > 0].sort_values(ascending=False)
print(missing_summary)
missing_rows = df[df.isnull().any(axis=1)]
print(missing_rows)

```

```python
df = df.dropna(subset=['email', 'What is your LinkedIn profile?', 'Job Title'])
```

```python
df['has_joined_event'] = df['has_joined_event'].str.lower()
df['has_joined_event'] = df['has_joined_event'].apply(lambda x: 'Yes' if x == 'yes' else 'No')
```

```python
cols = ['amount', 'amount_tax', 'amount_discount']
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
```

```python
df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')
df['created_date'] = df['created_at'].dt.date
df['created_time'] = df['created_at'].dt.time
```
